```python
from transformers import AutoTokenizer
from datasets import Dataset
import pandas as pd

# Load dataset hasil preprocessing
df = pd.read_csv("../data/processed/yt_clean.csv")

# Tokenizer IndoBERT
tokenizer = AutoTokenizer.from_pretrained("indobenchmark/indobert-base-p1")

def tokenize_function(example):
    return tokenizer(example["text"], truncation=True)

df["text"] = df["text"].fillna("").astype(str)
df["label"] = df["label"].astype(int)
dataset = Dataset.from_pandas(df)
```

```python
# Tokenisasi
tokenized = dataset.map(tokenize_function, batched=True)

# Split train & test
tokenized_split = tokenized.train_test_split(test_size=0.2)

train_dataset = tokenized_split["train"]
eval_dataset = tokenized_split["test"]

# Simpan dataset tokenized (opsional)
train_dataset.save_to_disk("../data/tokenized/train/")
eval_dataset.save_to_disk("../data/tokenized/eval/")
```
