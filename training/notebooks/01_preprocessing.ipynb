```python
import pandas as pd
import re

# Fungsi pembersihan teks
def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"#\w+", "", text)
    text = re.sub(r"[^a-zA-Z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
# Load dataset
df = pd.read_csv("../data/raw/yt_6500.csv")

# Tampilkan 20 data pertama
df.head(20)
```

```python
# Bersihkan teks
df["text"] = df["text"].apply(clean_text)

# Cek hasil
df.head(20)
```

```python
# Simpan hasil preprocess
df.to_csv("../data/processed/yt_clean.csv", index=False)
```
